```python
new_cols_dict ={
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_,kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'  
}
df.rename(new_cols_dict, axis = 1, inplace = True)
```

```python
df.humidity_percent = df.humidity_percent /100
```

```python
#creating season dictionary so that we can map the values
season_dict ={
    '0:0' : 'spring',
    '1.0' : 'summer',
    '2.0': 'autumn',
    '3.0' : 'winter'
}
#creating weather dictionary so that we can map the values
weather_dict ={
    '1.0' : 'clear',
    '2.0' : 'Scattered clouds',
    '3.0' : 'Broken clouds',
    '4.0' : 'Cloudy',
    '7.0' : 'Rain',
    '10.0' : 'Rain with thunderstorm',
    '26.0' : 'Snowfall'
}
#changing the season column datatype to string
df.season = df.season.astype('str')
#mapping thevalues 0-3 to the written seasons in season dictionary
df.season = df.season.map(season_dict)

#changing the weather column datatype to string
df.weather = df.weather.astype('str')
#mapping the weather column values as in weather dictionary
df.weather = df.weather.map(weather_dict)
```

```python
#writing the final dataframe to an excel file so that we will use it in our tableau visualization
df.to_excel('london_final.xlsx', sheet_name ='Data')
```
